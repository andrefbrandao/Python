```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import table
```

```python

industria_formal     = pd.read_excel("gini por setor.xlsx", sheet_name = 11, index_col = 0).round(4)
industria_informal   = pd.read_excel("gini por setor.xlsx", sheet_name = 12, index_col = 0).round(4)
agricultura_formal   = pd.read_excel("gini por setor.xlsx", sheet_name = 13, index_col = 0).round(4)
agricultura_informal = pd.read_excel("gini por setor.xlsx", sheet_name = 14, index_col = 0).round(4)
serviços_formal      = pd.read_excel("gini por setor.xlsx", sheet_name = 15, index_col = 0).round(4)
serviços_informal    = pd.read_excel("gini por setor.xlsx", sheet_name = 16, index_col = 0).round(4)
setor_publico        = pd.read_excel("gini por setor.xlsx", sheet_name = 17, index_col = 0).round(4)

```

```python
industria_formal
```

```python
fig, axs = plt.subplots(4,2, figsize=(15, 15))


# Tabela 1
axs[0, 0].axis('off')  # Não exibe eixos
tbl1 = table(axs[0, 0], industria_formal, loc='center')
tbl1.auto_set_font_size(False)
tbl1.set_fontsize(10)
axs[0, 0].set_title('Indústria Formal')

# Tabela 2
axs[0, 1].axis('off')
tbl2 = table(axs[0, 1], industria_informal, loc='center')
tbl2.auto_set_font_size(False)
tbl2.set_fontsize(10)
axs[0, 1].set_title('Indústria Informal')

# Tabela 3
axs[1, 0].axis('off')
tbl3 = table(axs[1, 0], agricultura_formal, loc='center')
tbl3.auto_set_font_size(False)
tbl3.set_fontsize(10)
axs[1, 0].set_title('Agricultura Formal')

# Tabela 4
axs[1, 1].axis('off')
tbl4 = table(axs[1, 1], agricultura_informal, loc='center')
tbl4.auto_set_font_size(False)
tbl4.set_fontsize(10)
axs[1, 1].set_title('Agricultura Informal')

# Tabela 5
axs[2, 0].axis('off')
tbl5 = table(axs[2, 0], serviços_formal, loc='center')
tbl5.auto_set_font_size(False)
tbl5.set_fontsize(10)
axs[2, 0].set_title('Serviços Formal')

# Tabela 6
axs[2, 1].axis('off')
tbl6 = table(axs[2, 1], serviços_informal, loc='center')
tbl6.auto_set_font_size(False)
tbl6.set_fontsize(10)
axs[2, 1].set_title('Serviços Informal')

# Tabela 7
axs[3, 0].axis('off')
tbl7 = table(axs[3, 0], serviços_formal, loc='center')
tbl7.auto_set_font_size(False)
tbl7.set_fontsize(10)
axs[3, 0].set_title('Setor Público')

axs[3, 1].axis('off')
plt.savefig('tabelas gini e derivados por setor.png')
plt.show()

```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4, figsize=(20, 8))  # 2 linhas, 4 colunas


# gráfico 1
ax1.plot(industria_formal['GINI'], label='Gini')
ax1.plot(industria_formal['CC'], label='Coeficiente de Concentração')
ax1.set_title('Indústria', fontsize=14)
ax1.set_ylabel('Gini e CC - Setor Formal', fontsize=14)
ax1.set_ylim(-0.1, 0.3) 
ax1.legend()
ax1.grid()
ax1.axhline(0, color='black', linestyle='--', linewidth=1)

# gráfico 2
ax2.plot(agricultura_formal['GINI'], label='Gini')
ax2.plot(agricultura_formal['CC'], label='Coeficiente de Concentração')
ax2.set_title('Agricultura', fontsize=14)
ax2.set_ylim(-0.1, 0.3) 
ax2.legend()
ax2.grid()
ax2.axhline(0, color='black', linestyle='--', linewidth=1)


# gráfico 3
ax3.plot(serviços_formal['GINI'], label='Gini')
ax3.plot(serviços_formal['CC'], label='Coeficiente de Concentração')
ax3.set_title('Serviços', fontsize=14)
ax3.set_ylim(-0.1, 0.3) 
ax3.legend()
ax3.grid()
ax3.axhline(0, color='black', linestyle='--', linewidth=1)


# gráfico 4
ax4.plot(setor_publico['GINI'], label='Gini')
ax4.plot(setor_publico['CC'], label='Coeficiente de Concentração')
ax4.set_title('Setor Público', fontsize=14)
ax4.set_ylim(-0.1, 0.3) 
ax4.legend()
ax4.grid()
ax4.axhline(0, color='black', linestyle='--', linewidth=1)

# gráfico 5
ax5.plot(industria_informal['GINI'], label='Gini')
ax5.plot(industria_informal['CC'], label='Coeficiente de Concentração')
ax5.set_ylim(-0.1, 0.3) 
ax5.set_ylabel('Gini e CC - Setor Informal', fontsize=14)
ax5.legend()
ax5.grid()
ax5.axhline(0, color='black', linestyle='--', linewidth=1)


# gráfico 6
ax6.plot(agricultura_informal['GINI'], label='Gini')
ax6.plot(agricultura_informal['CC'], label='Coeficiente de Concentração')
ax6.set_ylim(-0.1, 0.3) 
ax6.legend()
ax6.grid()
ax6.axhline(0, color='black', linestyle='--', linewidth=1)

# gráfico 7
ax7.plot(serviços_informal['GINI'], label='Gini')
ax7.plot(serviços_informal['CC'], label='Coeficiente de Concentração')
ax7.set_ylim(-0.1, 0.3) 
ax7.legend()
ax7.grid()
ax7.axhline(0, color='black', linestyle='--', linewidth=1)

ax8.axis('off')



plt.subplots_adjust(hspace=0.2, wspace=0.4)

plt.savefig('grafico gini e cc por setor.png')
```
